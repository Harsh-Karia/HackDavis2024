```python
import json

with open(r'hackdavis2024-default-rtdb-export.json', 'r') as f:
    d = json.load(f)
d 
```

```python
hari = {key:value for key, value in d.items() if "4:52" in key}
aarav = {key:value for key, value in d.items() if "4:55" in key}
harsh = {key:value for key, value in d.items() if "4:57" in key}
print(len(hari), len(aarav), len(harsh))
hari
```

```python
import json
from datetime import datetime
import pandas as pd

# Sample data in JSON format (replace with your actual JSON data)
data = hari

# Convert data to DataFrame
rows = []
for time_str, values in data.items():
    row = {
        'timestamp': datetime.strptime(time_str, '%H:%M:%S'),
        'epochTime': values['epochTime'],
        'off': values['off'],
        'on': values['on'],
        'difference': values['on'] - values['off']
    }
    rows.append(row)

df = pd.DataFrame(rows)

# Set timestamp as index
df.set_index('timestamp', inplace=True)

# Group by minute and calculate averages
result = df.resample('T').agg({
    'on': 'mean',
    'difference': 'mean'
})

result
```
